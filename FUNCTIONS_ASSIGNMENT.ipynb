# 3. What are the different ways to define and call a function in python?

```python
def only_numeric (a):
  n = []
  for i in a:
    if type(i) == int or type(i) == float:
      n.append(i)
      return n
```

```python
only_numeric([1, True, 'ajay', 3+5j])
```

# 7. What are the advantages of using generators over regular functions?

While regular functions use lots of space and memory, generator's yeild keyword is useful when working with large data sets, allowing for lazy evaluation and efficient use of memory.

```python
def square_number(n):
  for i in range(n):
    yield i**2
```

# 8. Use a lambda function in python to sort a list of tuple based on the second element of each tuple.

```python
subject_marks = [('english', 83), ('history', 76), ('geography', 87), ('maths', 66)]
subject_marks
```

# 9. Create a python program that uses 'filter()' to remove all the vowels from a given string.

```python
s = "Assignment"
vowels = ['a','e','i','o','u','A','E','I','O','U']
result = ""
for i in range(len(s)):
  if s[i] not in vowels:
    result = result + s[i]
print("\nAfter removing vowels: ", result)
```
